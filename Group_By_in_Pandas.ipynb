# Group By in pandas

```python
import pandas as pd
```

### Import the data

```python
x=pd.read_csv("C:\\Users\\Satadru Mukherjee\\Desktop\\iris_csv.csv") # Note Give \\ to take from directory
```

```python
x.head()
```

### Pandas dataframe.groupby() function is used to split the data into groups based on some criteria.

```python
ms=x.groupby('class')
```

### Get description of all the individual groups

```python
ms.describe().unstack(1)
```

### We use the function get_group() to find the entries contained in any of the groups.

```python
ms.get_group('Iris-virginica')       #Just write get_group & then the group , for which you want the information.
#You will be getting whole data of the group
```
